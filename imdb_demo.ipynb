```python
import selenium
import time
import random
```

```python
from selenium import webdriver
```

* FILM YILLARINI CEKMEK

```python
driver = webdriver.Firefox()
url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"

driver.get(url)
time.sleep(2)
elements = driver.find_elements_by_css_selector(".secondaryInfo")


for element in elements:
    print("***************")
    print(element.text)

driver.close()
```
